Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

Question 1: Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$

Question 2: Given the following data
\begin{gather}
\begin{bmatrix}
\begin{array}{c|c}
Age & Proportion \\
18-24 & 0.18 \\
25-34 & 0.23 \\
35-44 & 0.16 \\
45-64 & 0.27 \\
65-100 & 0.16 \\
\end{array}
\end{bmatrix}
\end{gather}

If 500 adults are sampled  randomly, find the probability rhat the sample contains 100 person between 18 and 24, 200 between, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64.. What is the expected value for to obtain a person in the 65 and above?

#### ANSWERS:

##### QUESTION 1:
To find the expected value $( E[Y_i] )$ of the random variable $( Y_i )$, we need to sum over all possible values of $( Y_i )$ weighted by their probabilities.

Given the joint probability distribution function $( P(y) )$, which represents the probability of observing a particular outcome vector $( y = (y_1, y_2, \ldots, y_k) )$, where $( y_i )$ represents the number of trials resulting in outcome $( i )$, we can find the expected value of $( Y_i )$ as follows:

\begin{equation}
E[Y_i] = \sum_{\text{all } y} y_i \cdot P(y)
\end{equation}

Where $ \sum_{\text{all } y} $ denotes the sum over all possible outcome vectors.

Given $( P(y) = \frac{n!}{y_1!y_2! \ldots y_k!} p_1^{y_1} p_2^{y_2} p_3^{y_3} \ldots p_k^{y_k} )$, we can substitute this into the expression for the expected value:

\begin{equation}
E[Y_i] = \sum_{\text{all } y} y_i \cdot \frac{n!}{y_1!y_2! \ldots y_k!} p_1^{y_1} p_2^{y_2} p_3^{y_3} \ldots p_k^{y_k}
\end{equation}

Now, notice that $( y_i )$ only contributes to the sum when $( y_i )$ is present, because $( y_i )$ counts the number of trials for which the outcome falls into cell $( i )$. So, we have:

\begin{equation}
E[Y_i] = \sum_{\text{all } y} y_i \cdot \frac{n!}{y_1!y_2! \ldots y_k!} p_1^{y_1} p_2^{y_2} p_3^{y_3} \ldots p_k^{y_k}
\end{equation}

Since the sum is over all possible outcome vectors $( y )$, and $( y_i )$ is a summation variable, this sum just counts how many times $( y_i )$ appears in all possible outcome vectors, which is simply $( n )$, the total number of trials. 

Therefore, we have:

\begin{equation}
E[Y_i] = n \cdot p_i
\end{equation}

This shows that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$


##### QUESTION 2:

If 500 adults are sampled randomly, we need to find the probability that the sample contains:
- 100 persons between 18 and 24,
- 200 persons between the ages of 25 and 34,
- 200 persons between the ages of 45 and 64.

We also need to find the expected value for obtaining a person in the 65 and above age group.

First, let's denote the number of people in each age group by $(Y_i )$ where $(i = 1, 2, 3, 4, 5 )$ corresponding to the age groups:
- $(Y_1 )$ for 18-24
- $(Y_2 )$ for 25-34
- $(Y_3 )$ for 35-44
- $(Y_4 )$ for 45-64
- $(Y_5 )$ for 65-100

The given proportions $(p_i )$ for each age group are:
\begin{equation}
p_1 = 0.18, 
p_2 = 0.23, 
p_3 = 0.16, 
p_4 = 0.27, 
p_5 = 0.16.
\end{equation}

The probability of observing a specific outcome vector $(y = (y_1, y_2, y_3, y_4, y_5) )$ is given by the multinomial distribution:
\begin{equation}
P(y) = \frac{n!}{y_1! y_2! y_3! y_4! y_5!} p_1^{y_1} p_2^{y_2} p_3^{y_3} p_4^{y_4} p_5^{y_5}
\end{equation}
where $(n = 500 )$ is the total sample size.

We need to find $(P(Y_1 = 100, Y_2 = 200, Y_4 = 200) )$, with $(Y_3 )$ and $(Y_5 )$ determined by the constraints:
\begin{equation}
Y_3 = 500 - (Y_1 + Y_2 + Y_4 + Y_5) = 500 - (100 + 200 + 200) = 0,
\end{equation}
\begin{equation}
Y_5 = 500 - (100 + 200 + 200) = 0.
\end{equation}

Substituting these values into the multinomial probability formula, we get:
\begin{equation}
P(Y_1 = 100, Y_2 = 200, Y_3 = 0, Y_4 = 200, Y_5 = 0) = \frac{500!}{100! 200! 0! 200! 0!} (0.18)^{100} (0.23)^{200} (0.16)^{0} (0.27)^{200} (0.16)^{0}
\end{equation}

The expected value for obtaining a person in the 65 and above age group, $(E[Y_5] )$, can be calculated using the formula for the expected value of a multinomial distribution:
\begin{equation}
E[Y_5] = n \cdot p_5 = 500 \cdot 0.16 = 80.
\end{equation}

Thus, the probability of the specified sample is:
\begin{equation}
P(Y_1 = 100, Y_2 = 200, Y_3 = 0, Y_4 = 200, Y_5 = 0) = \frac{500!}{100! 200! 0! 200! 0!} (0.18)^{100} (0.23)^{200} (0.27)^{200}
\end{equation}

And the expected value for obtaining a person in the 65 and above age group is:
\begin{equation}
E[Y_5] = 80.
\end{equation}

2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to q = $(1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$ , the number of successes observed during the $n$ trials.

Question 3: Experience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. #### ANSWER:

##### QUESTION 1:

For $( Y = k )$ (where $( k )$ successes occur in $( n )$ trials), there are $(\binom{n}{k})$ ways to choose which $( k )$ trials result in success out of $( n )$ trials.

The probability of any specific sequence of $( k )$ successes and $( n - k )$ failures is $( p^k q^{n-k} )$, given the independence of trials.

The probability of observing exactly $( k )$ successes in $( n )$ trials is the number of such sequences multiplied by the probability of any one sequence. Therefore,

\begin{equation}
P(Y = k) = \binom{n}{k} p^k q^{n-k}
\end{equation}

Since $q = 1 - p$, we can substitute this in the equation.

Therefore, the probability distribution of the experiment is:

\begin{equation}
P(Y = k) = \binom{n}{k} p^k (1 - p)^{n-k}
\end{equation}

for $( k = 0, 1, 2, \ldots, n )$.

##### QUESTION 2

To show that the expectation of the binomial distribution is $( E[Y] = np )$, we start with the definition of expectation for a discrete random variable. For a binomially distributed random variable $( Y \sim \text{Binomial}(n, p) )$, the expectation $( E[Y] )$ is given by:

\begin{equation}
E[Y] = \sum_{k=0}^{n} k \cdot P(Y = k)
\end{equation}

From the binomial distribution, we know that:

\begin{equation}
P(Y = k) = \binom{n}{k} p^k (1 - p)^{n - k}
\end{equation}

Substituting this into the expectation, we have:

\begin{equation}
E[Y] = \sum_{k=0}^{n} k \cdot \binom{n}{k} p^k (1 - p)^{n - k}
\end{equation}

To simplify this, we use the properties of binomial coefficients and algebraic manipulation. Note that:

\begin{equation}
k \cdot \binom{n}{k} = k \cdot \frac{n!}{k!(n-k)!} = \frac{n!}{(k-1)!(n-k)!} = n \cdot \frac{(n-1)!}{(k-1)!(n-k)!} = n \cdot \binom{n-1}{k-1}
\end{equation}

Thus,

\begin{equation}
E[Y] = \sum_{k=0}^{n} k \cdot \binom{n}{k} p^k (1 - p)^{n - k} = \sum_{k=1}^{n} n \cdot \binom{n-1}{k-1} p^k (1 - p)^{n - k}
\end{equation}

Note that for $( k = 0 )$, the term is zero because $( k \cdot P(Y = 0) = 0 )$. Therefore, we can adjust the sum to start from $( k = 1 )$.

We can factor out $( n \cdot p )$ from the sum:

\begin{equation}
E[Y] = n \cdot p \sum_{k=1}^{n} \binom{n-1}{k-1} p^{k-1} (1 - p)^{(n-1)-(k-1)}
\end{equation}

Notice that the summation now represents the binomial expansion of $( (p + (1 - p))^{n-1} )$, which equals $( 1 )$:

\begin{equation}
\sum_{k=1}^{n-1} \binom{n-1}{k-1} p^{k-1} (1 - p)^{(n-1)-(k-1)} = (p + (1 - p))^{n-1} = 1
\end{equation}

Thus,

\begin{equation}
E[Y] = n \cdot p \cdot 1 = np
\end{equation}

Therefore, we have shown that the expected value of a binomially distributed random variable $( Y )$ is:

\begin{equation}
E[Y] = np
\end{equation}

##### QUESTION 3

Suppose $( p )$ is the probability that a person with the illness recovers, and $( 1 - p )$ is the probability that they do not recover. Given that 30% of all persons afflicted by the illness recover, we have $( p = 0.30 )$ and $( 1 - p = 0.70 )$.

The scenario can be modeled using a binomial distribution. Let $( Y )$ be the number of people out of ten who recover after receiving the medication. We are interested in finding the probability that at least nine out of ten people recover.

Using the binomial probability formula:

\begin{equation}
P(Y = k) = \binom{10}{k} p^k (1 - p)^{10 - k}
\end{equation}

where $( k )$ represents the number of successes (recoveries) out of ten trials (people), and $( \binom{10}{k} )$ is the number of ways to choose $( k )$ successes out of 10 trials.

For $( k = 9 )$ and $( k = 10 )$, we have:

\begin{equation}
P(Y \geq 9) = P(Y = 9) + P(Y = 10)
\end{equation}

Substituting the values:

\begin{equation}
P(Y \geq 9) = \binom{10}{9} p^9 (1 - p)^{10 - 9} + \binom{10}{10} p^{10} (1 - p)^{10 - 10}
\end{equation}

Calculating each term:

\begin{equation}
P(Y \geq 9) = \binom{10}{9} (0.30)^9 (0.70)^{10 - 9} + \binom{10}{10} (0.30)^{10} (0.70)^{10 - 10}
\end{equation}

\begin{equation}
P(Y \geq 9) = \binom{10}{9} (0.30)^9 (0.70)^{1} + \binom{10}{10} (0.30)^{10} (0.70)^{0}
\end{equation}

\begin{equation}
P(Y \geq 9) = \binom{10}{9} (0.30)^9 (0.70) + \binom{10}{10} (0.30)^{10}
\end{equation}

Using the binomial coefficient:

\begin{equation}
\binom{10}{9} = 10, \quad \binom{10}{10} = 1
\end{equation}

Substituting these values:

\begin{equation}
P(Y \geq 9) = 10 (0.30)^9 (0.70) + 1 (0.30)^{10}
\end{equation}

Therefore, the probability that at least nine out of ten people recover after receiving the medication is:
\begin{equation}
P(Y \geq 9) = 0.0001436859
\end{equation}

P(Y \geq 1) = 1 - P(Y = 0)

Task 3: Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

Question 1: Find a general formula for the expected value of this distribution.

Question 2: A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.


#### ANSWER:

##### QUESTION 1:

To find the expected value $(E[Y])$ of the given probability distribution $(p(y) = \frac{\lambda^y}{y!} e^{-\lambda})$, we need to sum over all possible values of $(Y)$ weighted by their probabilities.

The expected value $(E[Y])$ is given by:

\begin{equation} E[Y] = \sum_{y=0}^{\infty} y \cdot p(y) \end{equation}

Substituting the given probability distribution $(p(y))$:

\begin{equation} E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} \end{equation}

\begin{equation} E[Y] = e^{-\lambda} \sum_{y=0}^{\infty} \frac{\lambda^y}{(y-1)!} \end{equation}

\begin{equation} E[Y] = e^{-\lambda} \left( \sum_{y=1}^{\infty} \frac{\lambda^y}{(y-1)!} + 0 \right) \end{equation}

Now, we observe that the sum resembles the Taylor series expansion of $( e^{\lambda} )$. Therefore, we rewrite it as:

\begin{equation} E[Y] = e^{-\lambda} \left( e^{\lambda} \right) \end{equation}

\begin{equation} E[Y] = e^{-\lambda + \lambda} \end{equation}

\begin{equation} E[Y] = e^0 \end{equation}

\begin{equation} {E[Y] = 1} \end{equation}

So, the general formula for the expected value of this distribution is $( E[Y] = 1 )$.

##### QUESTION 2:

To solve this problem, we can model it using a Poisson distribution since the number of seedlings per square yard follows approximately a Poisson distribution with a mean of $( \lambda = 5 )$. 

Let $( Y )$ be the number of seedlings in a 1-square-yard sampling region. Then $( Y )$ follows a Poisson distribution with parameter $( \lambda = 5 )$. 

We are interested in finding the probability that none of the regions will contain seedlings. 

The probability mass function (PMF) of a Poisson distribution is given by:

\begin{equation} P(Y = y) = \frac{e^{-\lambda} \lambda^y}{y!} \end{equation}

Substituting $( \lambda = 5 )$, we get:

\begin{equation} P(Y = y) = \frac{e^{-5} 5^y}{y!} \end{equation}

Now, we want to find the probability that none of the regions will contain seedlings. Since each region is independent, we can use the probability mass function for a single region to calculate this probability.

For a single region, $( y = 0 )$, so:

\begin{equation} P(Y = 0) = \frac{e^{-5} 5^0}{0!} = e^{-5} \end{equation}

Now, we need to find the probability that none of the ten regions will contain seedlings. Since each region is independent, the probability that none of the regions will contain seedlings is the product of the individual probabilities for each region:

\begin{equation} P(\text{None of the regions contain seedlings}) = (e^{-5})^{10} = e^{-50} \end{equation}

So, the probability that none of the regions will contain seedlings is approximately $( e^{-50} )$.
